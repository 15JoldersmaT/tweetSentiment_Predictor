```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


#set keywords to look for
keywords = [
    "love",
    "joy",
    "peace",
    "trust",
    "fear",
    "surprise",
    "sadness",
    "disgust",
    "anger",
    "anticipation",
    "ecstasy",
    "despair",
    "rage",
    "anxiety",
    "grief",
    "elation",
    "remorse",
    "excitement",
    "frustration",
    "satisfaction",
    "guilt",
    "pride",
    "humiliation",
    "jealousy",
    "hope",
    "shame",
    "envy",
    "gratitude",
    "euphoria",
    "regret",
    "compassion",
    "astonishment",
    "disappointment",
    "desire",
    "embarrassment",
    "loneliness",
    "contentment",
    "dread",
    "amazement",
    "bewilderment",
    "infatuation",
    "rejection",
    "horror",
    "revulsion",
    "triumph",
    "bliss",
    "devastation"
]
data = pd.read_csv('tweets.csv')

data['Text'] = data['Text'].str.lower()
#data = data[data["emotion"].str.contains("disgust|joy")==True]


#for i in data['Text']:
#    ilist = i.split()
#    for i2 in ilist:
#        keywords.append(i2)

#remove duplicates
#keywords = list(set(keywords))

#how many words?
print (str(len(data)))

# add a coloumn for each unique word
for i in keywords:
    data[i] = data['Text'].apply(lambda x: i in x)
    


x = data[keywords].values
y = data['emotion'].values
```

```python
# Splitting dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
# Creating Decision Tree classifier object
clf = DecisionTreeClassifier(random_state=41, max_depth=6)

# Training Decision Tree classifier
clf = clf.fit(X_train, y_train)

# Predicting the response for test dataset
y_pred = clf.predict(X_test)

# Importing metrics for accuracy calculation
from sklearn import metrics

# Model Accuracy
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))


fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (30, 25), dpi=300)






# Create the plot
plot_tree(clf, 
          feature_names = keywords,  
          class_names=  list(set(y)),
          filled = True, 
          ax = axes);

# Save the figure
fig.savefig('decision_tree_tweets.png')
plt.show()
```
